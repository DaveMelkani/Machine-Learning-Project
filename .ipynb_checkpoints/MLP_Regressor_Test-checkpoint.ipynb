```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

#### Read in the data

```python
train_views = pd.read_csv("train_views.csv")
test_views = pd.read_csv("test_views.csv")
```

```python
train_stream = pd.read_csv("train_stream.csv")
test_stream = pd.read_csv("test_stream.csv")
```

#### Define features and label

```python
views_features = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
           'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo',
           'Duration_ms', 'Likes', 'Comments']
views_label = ['Views']
```

```python
stream_features = ['Danceability', 'Energy', 'Key', 'Loudness', 'Speechiness',
           'Acousticness', 'Instrumentalness', 'Liveness', 'Valence', 'Tempo',
           'Duration_ms', 'Likes', 'Comments']
stream_label = ['Stream']
```

#### Initialize MLP architecture with 1 hidden layer of 100 neurons

```python
views_mlp = MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000, learning_rate_init=0.001)
```

```python
stream_mlp = MLPRegressor(hidden_layer_sizes=(100,), max_iter=1000, learning_rate_init=0.001)
```

#### Train the MLPs

```python
views_mlp.fit(train_views[views_features], train_views[views_label].values.ravel())
```

```python
stream_mlp.fit(train_stream[stream_features], train_stream[stream_label].values.ravel())
```

#### Make predictions on the test data

```python
views_pred = views_mlp.predict(test_views[views_features])
```

```python
stream_pred = stream_mlp.predict(test_stream[stream_features])
```

#### Evaluate the performance of the MLPs

```python
views_mse = mean_squared_error(test_views[views_label], views_pred)
views_r2 = r2_score(test_views[views_label], views_pred)
views_rmse = np.sqrt(views_mse)
# print("Views MLP mean squared error: {:.2f}".format(views_mse))
print("Views MLP root mean squared error: {:.2f}".format(views_rmse))
print("Views MLP R-squared score: {:.2f}".format(views_r2))
```

```python
stream_mse = mean_squared_error(test_stream[stream_label], stream_pred)
stream_r2 = r2_score(test_stream[stream_label], stream_pred)
stream_rmse = np.sqrt(stream_mse)
# print("Stream MLP mean squared error: {:.2f}".format(stream_mse))
print("Stream MLP root mean squared error: {:.2f}".format(stream_rmse))
print("Stream MLP R-squared score: {:.2f}".format(stream_r2))
```

#### MLP Regressor for Views & Stream Data

```python
# Scatter plot for views data
plt.scatter(test_views[views_label], views_pred)
plt.xlabel('Actual Views')
plt.ylabel('Predicted Views')
plt.title('MLP Regressor for Views Data')
plt.show()

# Scatter plot for stream data
plt.scatter(test_stream[stream_label], stream_pred)
plt.xlabel('Actual Stream')
plt.ylabel('Predicted Stream')
plt.title('MLP Regressor for Stream Data')
plt.show()
```

#### Create a correlation matrix of the features in the dataset

```python
views_corr = train_views[views_features].corr()
```

```python
stream_corr = train_stream[stream_features].corr()
```

#### Create a heatmap of the correlation matrix

```python
plt.figure(figsize=(10, 8))
sns.heatmap(views_corr, annot=True, cmap="coolwarm")
plt.title("Views Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(stream_corr, annot=True, cmap="coolwarm")
plt.title("Stream Correlation Heatmap")
plt.show()
```
